# Representing harmonic mean

The harmonic mean has the form
\\[
h(x_1, \ldots x_n) = \frac{n}{\frac{1}{x_1} + \cdots + \frac{1}{x_n}}.
\\]

The simple harmonic mean $h(x_1,x_2)$ (watch the factor of 2!) can be represented with the second-order cone constraint
\\[
v \leq \sqrt{ (v-x_1)(v- x_2) } \implies v \leq \frac{1}{\frac{1}{x_1} + \frac{1}{x_2}}
\\]
or
\\[
\mathrm{SOC}(x_1 + x_2 - 2v, [x_1 - x_2, 2v])
\\]

```python
import pygraphviz as PG
import os
from contextlib import contextmanager
from IPython.display import Image, display, Latex

@contextmanager
def temp_filename():
    name = 'temp.png'
    try:
        yield name
    finally:
        os.remove(name)

def show_graph(tree, **kwargs):
    A = PG.AGraph(directed=True, strict=True)
    cones = [k for k in tree.keys() if not isleaf(k)]
    root = tree.keys()[0]
    
    seen = set()
    tovisit = [root]
    for v in tovisit:
        if v not in seen:
            seen.add(v)

            for child in tree[v]:
                tovisit.append(child)
                A.add_edge(latex_var(cones,v,**kwargs),latex_var(cones,child,**kwargs))

    # pygraphviz renders graphs in neato by default, 
    # so you need to specify dot as the layout engine
    A.layout(prog='dot')
    with temp_filename() as f:
        A.draw(f)
        display(Image(f))

```

```python
from collections import deque
n = 17
d = deque(str(i) for i in range(n))

A = PG.AGraph(directed=True, strict=True)

while len(d) > 1:
    a = d.popleft()
    b = d.popleft()
    c = ','.join([a,b])
    d.append(c)
    A.add_edge(c,a)
    A.add_edge(c,b)


A.layout(prog='dot')
with temp_filename() as f:
    A.draw(f)
    display(Image(f))
```

```python
import numpy as np

def ohm(x,y):
    return 1.0/(1.0/x+1.0/y)

def test_soc(w,x,y):
    return np.sqrt((x-y)**2 + (2*w)**2), x + y - 2*w

def test_gm(v,x,y):
    return v, np.sqrt((v-x)*(v-y))
```

```python
hm(1,3)/2
```

```python
test_soc(2*1.5,1,3)
```

```python
x,y = 1, 3
test_gm(ohm(x,y), x, y)
test_soc(ohm(x,y), x, y)
```
