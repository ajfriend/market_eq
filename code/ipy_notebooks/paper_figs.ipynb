```python
from market_eq.market import gen_market3
from market_eq.solvers import admm_solver, cvxpy_solver
import matplotlib.pyplot as plt

from cStringIO import StringIO
import sys

import re

class Capturing(list):
    def __enter__(self):
        self._stdout = sys.stdout
        sys.stdout = self._stringio = StringIO()
        return self
    def __exit__(self, *args):
        self.extend(self._stringio.getvalue().splitlines())
        sys.stdout = self._stdout


def result(output):
    m = re.search(r'Status: (.*)',"\n".join(output))
    solved = m.group(1) == 'Solved'
    m = re.search(r'Total solve time: (.*)s',"\n".join(output))
    time = float(m.group(1))
    return solved, time

def displeasure(xs,phi):
    ''' Return average and maximum relative displeasure'''
    s = 0.0
    m = 0.0
    count = 0.0
    for a, x in xs.iteritems():
        t = a.displeasure(x,phi)
        s += t
        m = max(m, t)
        count += 1
        
    return s/count, m

def experiment(n, k):
    with Capturing() as output:
        market = gen_market3(n, k)
        xs, phi = cvxpy_solver(market)

    xs, phi = market.polish_sol(xs,phi)
    
    return dict(zip('n k solved time avg_resid max_resid'.split(), (n,k) + result(output) + displeasure(xs, phi)))
```

# generating random markets
use two random permutations to generate interest and allocation.

```python
sizes = [10, 25, 50, 100, 200]
goods = 20
runs = 4

a = [experiment(n,goods) for n in sizes for i in xrange(runs)]
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 16})

fig, (ax1,ax2) = plt.subplots(1,2, sharex=True, figsize=(10,5))

x,y = zip(*((d['n'], d['time']) for d in a))
ax1.plot(x,y,'o',alpha=.5)
ax1.set_ylabel('solve time in seconds')
ax1.set_xlabel('# of agents')

x,y,z = zip(*((d['n'], d['avg_resid'], d['max_resid']) for d in a))
ax2.plot(x,y,'bo', label='average', alpha=.5)
ax2.plot(x,z,'ro', label='worst-case', alpha=.5)
ax2.set_ylabel('relative residuals')
ax2.set_xlabel('# of agents')
ax2.margins(x=.1)
ax2.legend(loc='best', fancybox=True, framealpha=0.5)

fig.tight_layout()
```

```python
def admm_disp(g):
    for xs,phi in g:
        xs = xs.copy()
        phi = phi.copy()
        #print xs, 'XXX\n'
        xs, phi = market.polish_sol(xs,phi)
        yield displeasure(xs,phi)

market = gen_market3(15, 6)
```

```python
g = admm_solver(market,steps=100)
a = list(admm_disp(g))
```

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 16})

fig, ax = plt.subplots(1,1, figsize=(5,5))
x,y = zip(*a)
ax.semilogy(x,'o',alpha=.5, label='average')
ax.semilogy(y,'o',alpha=.5, label='worst-case')
ax.set_ylabel('relative residuals')
ax.set_xlabel('iteration #')
ax.margins(x=.1)
ax.legend(loc='best', fancybox=True, framealpha=0.5)

fig.tight_layout()
fig.savefig('admm.pdf')
```
