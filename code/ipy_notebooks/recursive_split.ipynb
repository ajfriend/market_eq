# Representing the weighted geometric mean with second-order cones

The weighted geometric mean
$$
w_0 \leq \prod_{i=0}^{n-1} x_i^{p_i},
$$
where $p_i > 0$ and $\sum_i p_i = 1$ can be represented as a set of second order cone constraints when the powers $p_i$ are rational.
We will represent the original geometric mean as a collection of simple geometric mean inequalities of the form
$$
w \leq y^{1/2} z^{1/2},
$$
which happen to be representable as second-order cones.

For example, consider the representation of $w_0 \leq x_0^{1/4} x_1^{1/4} x_2^{1/2}$ given below.
We see that it can be represented with two simple geometric mean inequalities.
A graph is also given below to illustrate the composition of the inequalities.

```python
t = (1,6)
v = tuple(Fraction(n,sum(t)) for n in t)
tree = decompose(v)
display(latex_rep(tree))
show_graph(tree, var=True, powers=True)
```

# Algorithm
To represent a given weighted geometric cone inequality as a set of simple geometric cone inequalities, we will associated a
weighted geometric cone with the vector of its weights. That is, we will associate the inequality
$w \leq x_0^{1/4} x_1^{1/4} x_2^{1/2}$ with the vector $(1/4,1/4,1/2)$. We can use this vector notation to denote any weighted geometric mean inequality, as long as the weights sum to 1.
As a special case, any unit vector, will correspond to an $x$ variable. For example, $(0,0,1)$ would correspond to the inequality $w \leq x_2$. Note that the cones whose vectors are all zeros, except for exactly two entries which are both $1/2$, can be represented with a single second order cone constraint. For example, consider $(1/2,1/2,0)$.

Composing two weighted geo-mean cones in a third, simple geo-mean cone will half the two vectors and add them to result in the third. For example, if we compose $(0,0,1)$ with $(1/2,1/2,0)$ as in the figure above, we get an inequality with the weights
$$
(1/4,1/4,1/2) = 1/2(0,0,1) + 1/2(1/2,1/2,0).
$$

This composition rule is the basic idea of the decomposition algorithm we'll use. We start with an intial vector of weights,
$v_0$ and split it to find two children vectors $v_1$ and $v_2$ such that $v_0 = 1/2(v_1+v_2)$, with the constraint that each child vector must sum to 1. We then recurse down the tree, splitting the children vectors until we only have leaf nodes which can be represented by simple geometric mean inequalities, e.g., $(0,1/2,1/2)$. We do not need to duplicate nodes with the same vector signature, so we can re-use nodes which appear more than once to reduce the required number of inequalities.

The final set of inequalities are given by the same tree structure as demonstrated above; children nodes are composed in a simple geometric mean inequality to produce a set of inequalities which are equivalent to the weighted geometric mean corresponding to the root node.

However, there are many ways to split a node into two children. The algorithm below uses a particular heuristic which seem to work well: We load a single child with the most significant binary bits of the elements in the parent vector, and put the remaining binary bits in the other child. This heuristic seems to minimize the number of cones which are needed.

# Decomposition code

```python
from fractions import Fraction
from operator import itemgetter
import math
from collections import OrderedDict

def split(v):
    """ Take in a tuple, v, representing the weights of a weighted geometric cone and return two 'children'
    tuples which would compose to produce the parent geo-mean.
    """
    v = list(v)
    child = [[0]*len(v), [0]*len(v)]
    
    while sum(child[0]) != 1:
        # get the most significant bit in the parent tuple
        # find the largest element in the parent tuple
        i, elem = max(enumerate(v), key=itemgetter(1))
    
        # remove most significant bit from the parent tuple
        rem = Fraction(2)**int(math.floor(math.log(elem,2)))
        v[i] -= rem
        
        # shift the most sigificant bit up by one binary place
        # and add it to first child that it fits into
        for c in child:
            if sum(c) + 2*rem <= 1:
                c[i] += 2*rem
                break
    
    # when the first child sums to 1, add the remainder of v to second child
    child[1] = map(lambda x,y: 2*(x+y), child[1], v)
    
    return tuple(tuple(c) for c in child)

# a simpler algorithm 
def split2(v):
    """ Take in a tuple, v, representing the weights of a weighted geometric cone and return two 'children'
    tuples which would compose to produce the parent geo-mean.
    """
    print 'ooh baby'
    v = list(v)
    child = [[0]*len(v), [0]*len(v)]
    
    while sum(child[0]) < 1:
        # get the most significant bit in the parent tuple
        # find the largest element in the parent tuple
        i, elem = max(enumerate(v), key=itemgetter(1))
    
        # remove most significant bit from the parent tuple
        rem = Fraction(2)**int(math.floor(math.log(elem,2)))
        
        # shift the most sigificant bit up by one binary place
        # and add it to first child that it fits into
        child[0][i] += 2*rem
        v[i] -= rem
    
    # when the first child sums to 1, add the remainder of v to second child
    child[1] = map(lambda x,y: 2*(x+y), child[1], v)
    
    return tuple(tuple(c) for c in child)


# a simpler algorithm 
def split3(v0):
    """ Take in a tuple, v, representing the weights of a weighted geometric cone and return two 'children'
    tuples which would compose to produce the parent geo-mean.
    """
    v0 = list(v0)
    v1, v2 = [0]*len(v0), [0]*len(v0)
    
    while sum(v1) != 1:
        # get the most significant bit in the parent tuple
        # find the largest element in the parent tuple        
        i, elem = max(enumerate(v0), key=itemgetter(1))
    
        # remove most significant bit from the parent tuple
        rem = Fraction(2)**int(math.floor(math.log(elem,2)))
        
        # shift the most sigificant bit up by one binary place
        # and add it to first child that it fits into
        v1[i] += 2*rem
        v0[i] -= rem
    
    # when the first child sums to 1, add the remainder of v to second child
    v2 = map(lambda x,y: 2*(x+y), v2, v0)
    
    return tuple(v1), tuple(v2)


def decompose(v, split=split):
    """ Recursively split vectors corresponding to weighted geo-mean cones until the leaves are all
    simple, SOCP-representable cones. The original cone at the root will be represented by composing
    the nodes in the resulting tree.
    Keep a dictionary of node tuples mapping to that node's children.
    The dictionary will allow us to re-use nodes whose tuple we have already seen.
    We use an OrderedDict so that the root node is the first element of tree.keys().
    """
    tree = OrderedDict() # list children of each node
    vs = [tuple(v)]
    for v in vs:
        if not isleaf(v):
            tree[v] = split(v)
        else:
            tree[v] = []
        
        for c in tree[v]:
            if c not in tree:
                vs.append(c)
            
    return tree
```

# Plotting and printing code

```python
import pygraphviz as PG
import os
from contextlib import contextmanager
from IPython.display import Image, display, Latex

def isleaf(v):
    return sum(1 for e in v if e == 1) == 1

def prettyFrac(f):
    if f.numerator == 0:
        return "0"
    else:
        return "{}/{}".format(f.numerator, f.denominator)

def prettyTuple(t):
    return ", ".join(str(e) for e in t)

@contextmanager
def temp_filename():
    name = 'temp.png'
    try:
        yield name
    finally:
        os.remove(name)
        
def latex_var(cones, node, var=True, powers=True):
    s = ""
    if var:
        if node in cones:
            base = 'w_'
            ind = cones.index(node)
        else:
            base = 'x_'
            ind = node.index(1)
        
        fmt = "{%d}" if ind > 9 else '%d'
        s += base + fmt%ind
        
        if powers:
            s += ': '
        
    if powers:
        s += prettyTuple(node)
    
    return s

def show_graph(tree, **kwargs):
    A = PG.AGraph(directed=True, strict=True)
    cones = [k for k in tree.keys() if not isleaf(k)]
    root = tree.keys()[0]
    
    seen = set()
    tovisit = [root]
    for v in tovisit:
        if v not in seen:
            seen.add(v)

            for child in tree[v]:
                tovisit.append(child)
                A.add_edge(latex_var(cones,v,**kwargs),latex_var(cones,child,**kwargs))

    # pygraphviz renders graphs in neato by default, 
    # so you need to specify dot as the layout engine
    A.layout(prog='dot')
    with temp_filename() as f:
        A.draw(f)
        display(Image(f))
        
def latex_ineqs(tree):
    cones = [k for k in tree.keys() if not isleaf(k)]
    
    for cone in cones:
        s = "w_{%d}"%(cones.index(cone)) + r" &\leq "
        for child in tree[cone]:
            if child in cones:
                s += r"w_{%d}^{1/2} "%(cones.index(child))
            if isleaf(child):
                s += r"x_{%d}^{1/2} "%(child.index(1))
        yield s

def latex_rep(tree):
    root = tree.keys()[0]
    
    orig = r"w_0 \leq "
    for i, val in enumerate(root):
        orig += r"x_{%d}^{%s} "%(i, val)

    r = r"""
    The weighted geometric mean inequality
    \begin{align}
    %s
    \end{align}
    can be represented by the set of inequalities
    \begin{align}
    %s.
    \end{align}
    """%(orig, r"\\".join(latex_ineqs(tree)))
        
    return Latex(r)
```

```python
#This example seems to suggest that the ordering doesn't matter in making a tree with the minimum number of nodes

from itertools import permutations

for t in set(permutations([2,1])):
    v = tuple(Fraction(n,sum(t)) for n in t)
    tree = decompose(v,split=split3)

    show_graph(tree, var=False, powers=True)
    display(latex_rep(tree))
```
